```python
# SQLAlchemy
from sqlalchemy import create_engine
from sqlalchemy import inspect
# Pandas
import pandas as pd
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine

# Path to sqlite
#database_path ="sqlite:///hawaii.sqlite"
```

```python
# Create an engine that can talk to the database
engine = create_engine("sqlite:///hawaii.sqlite")
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
# Get the name of the table. 
inspector = inspect(engine)
inspector.get_table_names()
```

```python
Measurement=Base.classes.measurement
Station= Base.classes.station
```

```python
# Using the inspector to print the column names within the 'measurement' table and its types
columns = inspector.get_columns('measurement')
for column in columns:
    print(column)
```

```python
# Using the inspector to print the column names within the 'station' table and its types
columns = inspector.get_columns('station')
for column in columns:
    print(column)
```

```python
# Get the column names for the table
column_names = inspector.get_columns('measurement')
for column_name in column_names:
    print(column_name["name"])
```

```python
# Get the column names for the table
column_names = inspector.get_columns('station')
for column_name in column_names:
    print(column_name["name"])
```

```python
# Query All Records in the the Database
data = engine.execute("SELECT DISTINCT station name FROM measurement;")

for record in data:
    print(record)
```

```python
data = engine.execute("SELECT tobs FROM measurement;")

for record in data:
    print(record)
```

```python
# Query All Records in the the Database
data = engine.execute("SELECT station,name FROM station;")

for record in data:
    print(record)
```

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract
```

```python
session=Session(engine)
session.query(Measurement.date,Measurement.prcp).filter(extract('month',Measurement.date)==6)
results=[]
results=session.query(Measurement.date,Measurement.prcp)
print(results)

results=session.query(Measurement.date,Measurement.prcp).filter(extract('month',Measurement.date) == 6)

print(results)
results=session.query(Measurement.date,Measurement.prcp).filter(extract('month',Measurement.date) == 6).all()
results
```
